## 2. What is a Vector?

A vector can be interpreted in multiple equivalent ways.

From a physical perspective, it represents a quantity with both magnitudeand direction. <br>
Geometrically, it can be seen as a displacement from theorigin. <br>
In machine learning, vectors are commonly used to represent data
as ordered collections of numbers.


## 3. Dot Product

The dot product measures how much two vectors point in the same direction.<br>
If two vectors are aligned, their dot product is large.<br>
If they are orthogonal, the dot product is zero.

In machine learning, this idea is used to measure similarity between data points.

## 4. Vector Operations

Having understood what vectors represent, we now examine how they interact through basic operations.

For vectors $\mathbf{a}$ and $\mathbf{b}$,

Vector addition combinescorresponding components.<br>
Geometrically, this corresponds to placing the tail of one vectorat the head of the other.<br>
The addition operation is written as:

$$
\mathbf{a} + \mathbf{b}
$$

  Scalar multiplication scales the magnitude of a vector by the given scalar.<br>
  The multiplication operation is written as:

$$k*\mathbf{a}$$


The magnitude of a vector is measured using its norm.<br>
For a vector $\mathbf{a}$, the norm is written as:

$$
\|\mathbf{a}\|
$$

Dot Product is the component of one vector over another.
For vectors $\mathbf{a}$ and $\mathbf{b}$, Dot product is written as :

$$
\mathbf{a}\cdot\mathbf{b} = \|\mathbf{a}\|\|\mathbf{b}\|\cos\theta 
$$ 

where $\theta$ is the angle between the two vectors.

## 5. Verifciation through code

```python
import numpy as np

a = np.array([3, 4])
b = np.array([4, 3])

dot= np.dot(a, b)
cos_theta = dot/ (np.linalg.norm(a)*np.linalg.norm(b))
print(dot,cos_theta)
theta=np.degrees(np.arccos(cos_theta))
```

## 6. Visualisation

```python
import numpy as np
import matplotlib.pyplot as plt

#Here we define the vectors
a = np.array([3, 4])
b = np.array([4, 3])

# Plot the vectors
plt.figure(figsize=(5, 5))
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', label='a')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='red', label='b')
plt.quiver(3, 4, 0.84, -1.12, angles='xy', scale_units='xy', scale=1, color='green', label='a⋅b')


# Axes
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.grid(True)
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Geometric Interpretation of Dot Product")
plt.show()
```

## 7. Geometric Interpretation of the Dot Product

The vectors $\mathbf{a}$ and $\mathbf{b}$ originate from the same point and form a small angle between them.<br>
This visual alignment explains why their dot product is large but not equal to the product of their magnitudes.

The cosine of the angle between the vectors quantifies this alignment and forms the basis of cosine similarity.

This shows that the dot product is maximized when vectors are aligned and becomes zero when they are orthogonal.<br>
The numerical result confirms the geometric interpretation:<br>
vectors with similar directions produce a larger dot product.<br>


## 8. Cosine Similarity

The dot product depends on both the direction and magnitude of vectors.<br>
In many applications, we care only about how aligned two vectors are, independent of their length.

Cosine similarity removes the effect of magnitude by normalizing both vectors. It is defined as:

$$
\cos \theta
= \frac{\mathbf{a} \cdot \mathbf{b}}{\|\mathbf{a}\|\,\|\mathbf{b}\|}
$$


## 9. Projections and Orthogonality

Projection of $\mathbf{a}$ on $\mathbf{b}$ is the component of a that lies along $\mathbf{b}$ .<br>
the component perpendicular to b does not play any part in the projection.<br>
Projection is useful to measure similarity.

$$
\operatorname{proj}_{\mathbf{b}}(\mathbf{a})
= \frac{\mathbf{a} \cdot \mathbf{b}}{\|\mathbf{b}\|^2}\,\mathbf{b} = \|\mathbf{a}\|\cos\theta
$$

The numerator signals the alignment of projection while the denominator serves  the purpose of normalisation of the vector

```python
import numpy as np

a = np.array([3, 4])
b = np.array([4, 3])

proj_a_on_b = (np.dot(a, b) / np.dot(b, b)) * b

a, b, proj_a_on_b
print("a · b =", np.dot(a, b))
print("projection of a on b =", proj_a_on_b)
```

## 10. Orthogonality

Two vectors are orthogonal if their dot product is zero

Orthogonality means no projection

In ML: orthogonal features carry independent information, they have no dependency

```python
c = np.array([4, -3])  # orthogonal to [3,4]

print(np.dot(a, c)) # outputs 0
```

## 11. Observations

- Dot product depends on both magnitude and direction.
- Normalizing vectors removes the effect of scale.
- In high dimensions, many vectors tend to become nearly orthogonal.
